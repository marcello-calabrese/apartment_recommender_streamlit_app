### **Cleaning the Seattle airbnb listings.csv file Source:** https://www.kaggle.com/code/rdaldian/airbnb-content-based-recommendation-system/data?select=listings.csv 

```python
# import main libraries
import numpy as np
import pandas as pd
```

```python
# get the file into a dataframe from listing folder

df = pd.read_csv('listing/listings.csv')

df.head()
```

```python
# check the columns of the dataframe
df.columns
```

```python
# columns to get rid of
col = ['id', 'host_id', 'host_name', 'neighbourhood', 'latitude', 'longitude', 'room_type', 'minimum_nights', 'number_of_reviews', 'last_review', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365', 'notes', 'transit', 'thumbnail_url', 'medium_url', 'picture_url',
       'xl_picture_url', 'host_id', 'host_url', 'host_name', 'host_since',
       'host_location', 'host_about', 'host_response_time',
       'host_response_rate', 'host_acceptance_rate', 'host_is_superhost',
       'host_thumbnail_url', 'host_picture_url', 'host_neighbourhood',
       'host_listings_count', 'host_total_listings_count',
       'host_verifications', 'host_has_profile_pic', 'host_identity_verified',
       'street', 'neighbourhood', 'neighbourhood_cleansed',
       'neighbourhood_group_cleansed', 'city', 'state', 'zipcode', 'market', 'weekly_price', 'monthly_price', 'security_deposit',
       'cleaning_fee', 'guests_included', 'extra_people', 'minimum_nights',
       'maximum_nights', 'calendar_updated', 'has_availability',
       'availability_30', 'availability_60', 'availability_90',
       'availability_365', 'calendar_last_scraped', 'number_of_reviews',
       'first_review', 'last_review', 'review_scores_rating',
       'review_scores_accuracy', 'review_scores_cleanliness',
       'review_scores_checkin', 'review_scores_communication',
       'review_scores_location', 'review_scores_value', 'requires_license',
       'license', 'jurisdiction_names', 'instant_bookable',
       'cancellation_policy', 'require_guest_profile_picture',
       'require_guest_phone_verification', 'calculated_host_listings_count',
       'reviews_per_month']

# drop the columns
df.drop(col, axis=1, inplace=True)
df.head()
```

```python
# get rid of more columns

df.columns
```

```python
col2 = ['listing_url', 'scrape_id', 'last_scraped', 'summary', 'experiences_offered', 'smart_location', 'country_code', 'country', 'is_location_exact', 'beds',
       'bed_type', 'amenities', 'square_feet']

# get rid of the columns in col2
df.drop(col2, axis=1, inplace=True)

df.head(3)
```

```python
df.shape
```

```python
# check null values
df.isnull().sum()
```

```python
df.info()
```

```python
# fill null values of column space with 'unknown'
df['space'].fillna('unknown', inplace=True)
# fill null values of column neighborhood_overview with 'unknown'
#df['neighbourhood_overview'].fillna('unknown', inplace=True)
# fill null values of column property_type with mode
df['property_type'].fillna(df['property_type'].mode()[0], inplace=True)
# fill null values of column bathrooms with median
df['bathrooms'].fillna(df['bathrooms'].median(), inplace=True)

df.head(3)
```

```python
df['neighborhood_overview'].fillna('unknown', inplace=True)
df.head(2)
```

```python
df.isnull().sum()
```

```python
# save the dataframe to csv
df.to_csv('listing/listings_cleaned.csv', index=False)
```
